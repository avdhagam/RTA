```python
import pandas as pd

def missing_values_table(df):
    # Total missing values
    mis_val = df.isnull().sum()
    
    # Percentage of missing values
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    
    # Create a table with the results
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    
    # Rename the columns
    mis_val_table = mis_val_table.rename(columns={0: 'Missing Values', 1: '% of Total Values'})
    
    # Sort the table by percentage of missing descending
    mis_val_table = mis_val_table[mis_val_table.iloc[:, 1] != 0].sort_values('% of Total Values', ascending=False).round(1)
    
    # Print summary information
    print(f"Your selected dataframe has {df.shape[1]} columns.")
    print(f"There are {mis_val_table.shape[0]} columns that have missing values.")
    
    # Return the dataframe with missing information
    return mis_val_table

# Call the function on your DataFrame
missing_values_table(df)
```

#### Encoding

for features with a Lot of unique values, dont use one-hot encoding, try another encoding method instead. (like label/ordinal encoding)

```python
def ordinal_encoder(df, feats):
    for feat in feats:
        feat_val = list(np.arange(df[feat].nunique()))
        feat_key = list(df[feat].sort_values().unique())
        feat_dict = dict(zip(feat_key,feat_val))
        df[feat]=df[feat].map(feat_dict)
    return df
df = ordinal_encoder(df, df.drop("Accident_severity", axis=1))
df
```

```python
# Assuming "Accident_severity" is the target variable and you want to exclude it
features_to_encode = df.drop("Accident_severity", axis=1).columns
df = ordinal_encoder(df, features_to_encode)

corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(30, 30))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu"
                )
```

The target values are severely imabalanced so we need to upsample it using SMOTE

#### Upsampling

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

counter_before  = Counter(y_train)
# print("class distribution before SMOTE:", counter_before)
print("=========================")
for k,v in counter_before.items():
    per = 100*v/len(y_train)
    print(f"Class : {k}, n = {v} , ({per:.2f}%)")

 
oversample = SMOTE() #creates an instance of smote so we can adjust the parameters of SMOTE if needed

X_train , y_train = oversample.fit_resample(X_train,y_train)

print("=========================")



counter_after = Counter(y_train)
for k,v in counter_before.items():
    per = 100*v/len(y_train)
    print(f"Class : {k}, n = {v} , ({per:.2f}%)")
    
print("=========================")

print("upsampled data shape", X_train.shape,y_train.shape)
```

```python
## Building an ML model
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from catboost import CatBoostClassifier

models = {"xgb":XGBClassifier(),
          "rf":RandomForestClassifier(),
          "extree":ExtraTreesClassifier(),
          "catboost":CatBoostClassifier(verbose=False)}
# Create a function to fit and score

def fit_and_score(models,X_train,X_test,y_train,y_test):
    # Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    for name, model in models.items():
        # fit the model
        model.fit(X_train,y_train)
        # Evaluate the model and append 
        model_scores[name] = model.score(X_test,y_test)
    return model_scores
```

```python
model_scores = (fit_and_score(models,X_train,X_test,y_train,y_test))
model_scores
```

## Model Comparision

```python
model_compare = pd.DataFrame(model_scores,index=["accuracy"])
model_compare
```

```python
model_compare.T.plot(kind="bar");
```

## Hyperparameter Tuning

Now we'll try to improve the scores of all three with the help of GridSearchCV and RandomizedSearchCV

#### XGBoost

First we'll use GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Define the XGBoost classifier
xgb_model = XGBClassifier()

param_grid = {
    'learning_rate': [0.01, 0.02],
    'n_estimators': [300,600],
    'max_depth': [4,5],
    'subsample': [0.8,0.6],
}

gs_xgboost = GridSearchCV(xgb_model,
                          param_grid = param_grid,
                          cv=5,
                          verbose=True)
gs_xgboost.fit(X_test,y_test)
```

```python
gs_xgboost.best_params_
```

```python
gs_xgboost.score(X_test,y_test)
```

Now, using RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

# Define the XGBoost classifier
rs_XGB_classifier = XGBClassifier()

param_dist = {
    'learning_rate': [0.1, 0.2, 0.3],  # Increase the learning rate
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5],
    'subsample': [1.0],
}


rs_XGBoost = RandomizedSearchCV(rs_XGB_classifier,
                                param_grid,
                                cv=5,
                                verbose=True)
rs_XGBoost.fit(X_test,y_test)
```
